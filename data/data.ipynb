```python
df_comp = pd.read_csv('competitions.csv', sep=',')
```

```python
df = pd.merge(df_app, df_comp, on='competition_id', how='left')
```

```python
df.head().T
```

```python
df.columns
```

```python
cols_to_use = ['player_id', 'date', 'player_name', 'competition_id',
       'yellow_cards', 'red_cards', 'goals', 'assists', 'minutes_played',
       'name', 'sub_type', 'type', 'country_name']
```

```python
df = df[cols_to_use]
```

```python
df = df[
            (df.type == 'domestic_league') & 
            (df.sub_type == 'first_tier')
    ]
```

```python
countries_to_keep = ['England', 'Spain', 'Germany', 'Italy', 'France']
```

```python
df = df[df.country_name.isin(countries_to_keep)]
```
